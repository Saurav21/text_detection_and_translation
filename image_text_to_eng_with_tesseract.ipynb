```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import pytesseract
```

```python
img = cv2.imread("/content/sentence.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
value, threshold = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

cv2_imshow(threshold)
print(value)
```

```python
invert = 255 - threshold

cv2_imshow(invert)
```

```python
!mkdir tessdata
!wget -O ./tessdata/por.traineddata https://github.com/tesseract-ocr/tessdata/blob/main/por.traineddata?raw=true
```

```python
config_tesseract = '--tessdata-dir tessdata'
text = pytesseract.image_to_string(invert, lang = 'por', config=config_tesseract)
print(text)
```

```python
import goslate
```

```python
gs = goslate.Goslate()
translate = gs.translate(text, "en")
```
